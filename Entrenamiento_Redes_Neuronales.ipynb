### Función Escalonada
funciona como un simple interruptor de "encendido y apagado" para la neurona.

Si la suma de la información que recibe la neurona supera un cierto límite (umbral), la función dice "¡Sí!" y arroja un 1 (la neurona se activa y pasa la señal). Si la información no alcanza ese límite, dice "¡No!" y arroja un 0 (la neurona se queda inactiva y bloquea la señal).

Básicamente, toma una decisión binaria estricta: es un todo o nada.

wx <= Umbral ➔ y = 0
wx > Umbral ➔ y = 1 
- f(wx) ➔ y = {0,1}

### Función Sigmoide 
- Los valores muy grandes saturan en 1. los valores muy bajos saturan en 0. 
- Este ofrece una "probabilidad" expresada en valores entre 0 y 1. 

formula:
f(x) = σ(x) = 1 / 1 - e^-x

### Función Tangente Hiperbólica 
- Esta función es similar a la sigmoide
- transforma cualquier valor que recible y lo comprime en un rango muy estricto ente -1 y 1 
- Este permite ver claro los estímulos positivos y negativos 
- Al estar centrada en cero, facilita que la red neurnal aprenda más rapido y de una manera super estable. 

formula:
f(x) = tanh(x) = (e^x - e^-e)/(e^x - e^-e)

### Función ReLU
- Esta funcion si el valor es negativo lo convierte en 0 y si es positivo lo deja tal y como está
- Es super rapida tras que es una operacion matemática super sencilla permite que las RNP entrenen a una velocidad mucho mayor. 
- Es una unidad Rectificada Linea Función. 
formula: 
f(x) = { 0 for x < 0 } 
         x for x >= 0

### Funciones de Coste 
Objetivo: Tenemos que definir una función de coste del error para saber como poder ptimizar los parametros. 

Entonces: ¿Qué función de coste podemos utilizar? 

La función de coste es como nuestro medidor de errores. Es una formila que nos dice que tan equivocada está la red neuronal en nuestras predicciones.
como tal el objetivo de todo el entrenamiento es simple, hacer que ese numero baje hasta ser lo mas cercano a cero posible

pongo un ejemplo para entenderlo mejor, es como que estoy intentando aprender a lanzar dardos al centro de la diana, pero con los ojos vendados. Lanzo el primer dardo y alguien me dice que desvié 50 centimetros, ahora esos "50"cm es mi función de coste. 

Sabiendo esto, ajusto mi postura y lanzo otra vez y ahora me dicen que desvié a 10 cm, como mi funcion de coste bajó bajó significa que estoy aprendiendo y acercandome al objetivo.

### BackPropagation

Este tema de backPropagation es la forma en la que una red neuronal aprende de sus errores en si, porque primero la red hace una prediccion normal donde a eso se le llama propagación hacia delante, luego de eso compara ese resultado con el valor real y se calcula que tan equivocado estuvo la red neuronal (predicción), después de eso se viene lo más importante... este error se regresa por toda la red desde la salida hasta la capas inciales, para ver que pesos fueron los responsables del fallo. Después de ese fallo la red ajusta esos perso usando derivadas para reducir el error de la siguiente iteración
