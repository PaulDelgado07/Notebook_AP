### Redes Neuronales Profundas
Objetivo: Comprender y aplicar los fundamentos de las redes neuronales profundas, desde el perceptrón multicapa hasta las arquitecturas recurrentes y el mecanismo de atención Transformer, utilizando TensorFlow y PyTorch.

Resultados de aprendizaje
Al terminar la Unidad el estudiante tendrá la capacidad de realizar lo siguiente:

- Explica los conceptos fundamentales del aprendizaje profundo, diferenciándolo del aprendizaje automático tradicional, e identifica las características principales de los frameworks TensorFlow y PyTorch, describiendo el funcionamiento del perceptrón simple y el perceptrón multicapa (MLP) como base de las arquitecturas profundas.
- Describe el proceso de entrenamiento de redes neuronales, detallando el papel de las funciones de activación (Sigmoid, Tanh, ReLU, Softmax), el algoritmo de backpropagation basado en la regla de la cadena, y el funcionamiento de los optimizadores SGD y Adam, así como las técnicas de regularización Dropout y Batch Normalization para prevenir el sobreajuste.
- Analiza la arquitectura de las redes neuronales recurrentes (RNN), identificando el problema del gradiente desvaneciente en secuencias largas y explicando cómo las variantes LSTM y GRU lo resuelven mediante mecanismos de puertas que controlan el flujo de información a lo largo de la secuencia.
- Describe el mecanismo de atención y la arquitectura Transformer, explicando los conceptos de Query, Key y Value, el funcionamiento de Self-Attention y Multi-Head Attention, y reconociendo sus aplicaciones en procesamiento de lenguaje natural, visión por computadora y modelos multimodales.
