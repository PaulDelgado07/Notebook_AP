2. Ahora encadenamos los 16 filtros se juntan o se ensamblan en un mapa de características de 200px, 200px, 16maps y la segunda capa que también es un filtro que dará como resultado 32 filtros más.

### Aspectos importantes de las CNN

#### Stride
El stride es el salto que da el filtro cuando recorre la imagen.

- stride = 1 → recorre pixel por pixel (más detalle)
- stride = 2 → salta de 2 en 2 (menos detalle, más rápido)

Mientras más grande el stride, más pequeña será la salida.

---

#### Padding
El padding consiste en agregar bordes (generalmente ceros) alrededor de la imagen.

Sirve para:
- no perder información en los bordes
- controlar el tamaño de la salida

---

#### Convolución
Es la operación principal de las CNN.

Un filtro recorre la imagen y extrae características como:
- bordes
- texturas
- patrones

---

#### Pooling Layer
Sirve para reducir el tamaño de la imagen manteniendo la información más importante.

Tipos:
- Max Pooling → toma el valor máximo
- Average Pooling → calcula el promedio

Ayuda a:
- reducir el costo computacional
- evitar overfitting
- hacer la red más eficiente

#### Activation Function
Se aplica después de la convolución.

La más común es ReLU:
- convierte valores negativos en 0
- permite que la red aprenda mejor

#### Flatten
Convierte la matriz en un vector.

Se usa para pasar a las capas finales de la red.

#### Fully Connected
Es la parte final de la red.

Aquí se realiza la clasificación final (por ejemplo: perro, gato, etc.).

### Data Augmentation

El data augmentation es una técnica que se usa para generar más datos a partir de las imágenes que ya tenemos, sin necesidad de recolectar nuevas. Básicamente, se aplican transformaciones como rotaciones, cambios de escala, volteos o ajustes de brillo, para que el modelo vea diferentes versiones de la misma imagen.

Esto ayuda a que la red no memorice los datos (overfitting) y aprenda mejor a generalizar. Por ejemplo, si tengo una imagen de un gato, puedo rotarla o invertirla horizontalmente y sigue siendo un gato, pero ahora el modelo aprende a reconocerlo en diferentes posiciones.

### Transfer Learning

El transfer learning es una técnica en la que se utiliza un modelo que ya fue entrenado previamente con grandes cantidades de datos, y se adapta para resolver un nuevo problema. En lugar de entrenar una red desde cero, se aprovecha el conocimiento que el modelo ya aprendió.

Por ejemplo, un modelo entrenado con millones de imágenes puede reutilizarse para clasificar otro tipo de imágenes, ajustando solo las últimas capas. Esto permite ahorrar tiempo, reducir el uso de datos y obtener buenos resultados incluso con datasets pequeños.
